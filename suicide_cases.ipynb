Import suicide rates reported by WHO

```python
owid_suicide_rates = pd.read_csv('../MindScape/data/suicide-rates-by-age-detailed-who.csv', encoding = 'utf-8')
```

```python
owid_suicide_rates.head()
```

Clean dataset and rename columns

```python
owid_suicide_rates.rename(columns={
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 15-24 years of age": "rate for 15-24 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 25-34 years of age": "rate for 25-34 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 35-44 years of age": "rate for 35-44 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 45-54 years": "rate for 45-54 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 55-64 years": "rate for 55-64 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 65-74 years": "rate for 65-74 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 75-84 years": "rate for 75-84 years",
                    "Crude suicide rates (per 100 000 population) - Sex: both sexes - Age group: 85+ years of age" : "rate for 85+ years"
                            }, 
                    inplace=True)
```

```python
column_names = [
    "rate for 15-24 years",
    "rate for 25-34 years",
    "rate for 35-44 years",
    "rate for 45-54 years",
    "rate for 55-64 years",
    "rate for 65-74 years",
    "rate for 75-84 years",
    "rate for 85+ years"
]

owid_suicide_rates['total_rate'] = owid_suicide_rates[column_names].sum(axis=1)
```

Export cleaned dataset as CSV file 

```python
owid_suicide_rates.to_csv('owid_suicide.csv', index=False) 
```
